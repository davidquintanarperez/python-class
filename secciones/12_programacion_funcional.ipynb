## Inmutabilidad
¿Qué es mutabilidad?
Un dato mutable puede cambiar en el futuro. (Ejemplo: tuplas en Python)


### ¿Qué es inmutabilidad?
Un dato inmutable NO cambia en el futuro.


### ¿Por qué es importante?
En programación funcional, evitaremos SIEMPRE los datos mutables (de ser posible).


La razón principal es que la mutabilidad va a dificultar razonar sobre el código.


¿Has visto "datos mutables" en una ecuación matemática?


### ¿No es esto... ineficiente?
Si y no.


Si, porque hay más uso de memoria.


No, porque simplifica mucha carga cognitiva en un sistema.


Solo en caso de optimización
La mutabilidad debería usarse cuando la eficiencia que ganamos agregándola es mayor que la complejidad cognitiva que se genera en el código.


Es decir, en muy pocos casos.

## Funciones totales y parciales
### Función total
Es una función que puede dar respuesta a todos los elementos de su dominio.


Es decir, que sin importar la entrada, la función será capaz de responder apropiadamente.


### Función Parcial
Una función que puede dar respuesta solo a algunos elementos de su dominio y a otros no.


Hay que tener cuidado con funciones que no están bien definidas, o que son inseguras de usar por no ser totales.


## Evaluación perezosa (Lazy)
### ¿En qué consiste?
Es una manera para evitar la evaluación de un valor hasta que sea utilizado.


Puede ser útil en casos donde no estemos seguros del orden en que algo se evaluara, o cuando queremos optimizar el tiempo de inicialización de muchos datos.


## Razonamiento con tipos
### Tipados vs. dinámicos
Los tipos de datos son muy importantes al momento de crear sistemas que escalen.


Los tipos mantienen la *consistencia* de un sistema, y se vuelven una manera de probarlo.


### Sistema de tipos
Los tipos son una parte crítica para hacer abstracciones. Se vuelven una manera de *documentar* las posibilidades que tenemos.


### Proposiciones y Pruebas
Los tipos, son proposiciones.
Los programas, son pruebas.
Nuestros programas son como una gran función matemática.


El compilador se convierte en un razonador que *prueba nuestro software* con base a en las proposiciones que le damos.


### Tipos vs String
Es un error usar String en vez de tipos, ya que *el compilador no puede probar nada* cuando hay cambios.


## Recursión
Una manera de abordar los problemas donde una función se llama a sí misma.


Es la manera clásica de programar en lenguajes funcionales.


## Tail recursion
Es una pequeña modificación sobre un algoritmo recursivo con el objetivo de optimizar su ejecución.


Lo que se busca es no llenar de stackcon llamadas, sino realizar la operación que queremos y pasar su acumulación al siguiente paso.
