## Recursion and tail recursion

```python
# 3!
def factorial(number):
    if number == 0:
        print("end")
        return 1
    else:
        print("Calculated {}".format(number))
        return number * factorial(number-1)
```

```python
print(factorial(3))
```

```python
# 3!
def factorial(number, result = 1):
    if number == 0:
        print("end")
        return result
    else:
        print("Calculated {}, result {}".format(number, result))
        return factorial(number-1, number*result)
```

```python
print(factorial(3))
```
